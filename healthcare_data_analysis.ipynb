```python
df.drop_duplicates(keep='first',inplace=True)
```

```python
df.shape
```

```python
cat_cols = []
num_cols = []
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        cat_cols.append(col)
    else:
        num_cols.append(col)
```

```python
cat_cols
```

```python
num_cols
```

```python
for col in num_cols[0:-1]:
    plt.figure(figsize=(20,2))
    sns.boxplot(x=df[col])
    plt.title(col)
```

```python
low_cat_cols = ['Gender','Blood Type','Medical Condition','Insurance Provider','Admission Type','Medication','Test Results']
```

```python
for col in low_cat_cols:
    plt.figure(figsize=(20,10))
    g = sns.countplot(x=col,data=df)
    g.set_xticklabels(g.get_xticklabels(),rotation=45,horizontalalignment='right')
    g.bar_label(g.containers[0])
    plt.title(col)
```
